```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

# for surface plot
from matplotlib import cm
from matplotlib.ticker import LinearLocator
```

```python
contrails1 = pd.read_csv('Extracted Data/contrails 1.csv')
# contrails1.head()
contrails1.set_index('Timesteps', inplace=True)
# contrails1.sort_values(by=["Percentage"], ascending=False)
contrails1
```

```python
contrails2 = pd.read_csv('Extracted Data/contrails 2.csv')
# contrails2.head()
contrails2.set_index('Timesteps', inplace=True)
# contrails2.sort_values(by=["Percentage"], ascending=False)
contrails2
```

```python
contrails3 = pd.read_csv('Extracted Data/contrails 3.csv')
# contrails3.head()
contrails3.set_index('Timesteps', inplace=True)
# contrails3.sort_values(by=["Ice"], ascending=False)
contrails3
```

```python
fig, axs = plt.subplots(3, 3,figsize=(15,15))

axs[0, 0].plot(contrails1.TotalParticles, label = "Contrails 1")
axs[0, 0].plot(contrails2.TotalParticles, label = "Contrails 2")
axs[0, 0].plot(contrails3.TotalParticles, label = "Contrails 3")
# Set the x axis label of the current axis.
axs[0, 0].set_xlabel('Time Steps')
# Set the y axis label of the current axis.
axs[0, 0].set_ylabel('Number of Particles')
# Set a title 
axs[0, 0].set_title('Total Number of Particles')

axs[0,1].plot(contrails1.TotalIcePercentage, label = "Contrails 1")
axs[0,1].plot(contrails2.TotalIcePercentage, label = "Contrails 2")
axs[0,1].plot(contrails3.TotalIcePercentage, label = "Contrails 3")
# Set the x axis label of the current axis.
axs[0,1].set_xlabel('Time Steps')
# Set the y axis label of the current axis.
axs[0,1].set_ylabel('Percentage')
# Set a title 
axs[0,1].set_title('Total Ice Percentage')


# plt.set_title('Total Ice Percentange')
axs[0,2].plot(contrails1.newIcePercentage, label = "Contrails 1")
axs[0,2].plot(contrails2.newIcePercentage, label = "Contrails 2")
axs[0,2].plot(contrails3.newIcePercentage, label = "Contrails 3")
# Set the x axis label of the current axis.
axs[0,2].set_xlabel('Time Steps')
# Set the y axis label of the current axis.
axs[0,2].set_ylabel('Percentage')
# Set a title 
axs[0,2].set_title('New Ice Percentage')


axs[1,0].plot(contrails1.Temp, label = "Contrails 1")
axs[1,0].plot(contrails2.Temp, label = "Contrails 2")
axs[1,0].plot(contrails3.Temp, label = "Contrails 3")
# Set the x axis label of the current axis.
axs[1,0].set_xlabel('Time Steps')
# Set the y axis label of the current axis.
axs[1,0].set_ylabel('Temperature')
# Set a title 
axs[1,0].set_title('Mean Temperature')


axs[1,1].plot(contrails1.IceVolume, label = "Contrails 1")
axs[1,1].plot(contrails2.IceVolume, label = "Contrails 2")
axs[1,1].plot(contrails3.IceVolume, label = "Contrails 3")
# axs[0,1]Set the x axis label of the current axis.
axs[1,1].set_xlabel('Time Steps')
# Set the y axis label of the current axis.
axs[1,1].set_ylabel('Ice Volume')
# Set a title 
axs[1,1].set_title('Total Ice Volume')


axs[1,2].plot(contrails1.Ice, label = "Contrails 1")
axs[1,2].plot(contrails2.Ice, label = "Contrails 2")
axs[1,2].plot(contrails3.Ice, label = "Contrails 3")
# Set the x axis label of the current axis.
axs[1,2].set_xlabel('Time Steps')
# Set the y axis label of the current axis.
axs[1,2].set_ylabel('Ice')
# Set a title 
axs[1,2].set_title('Total Number of Ice Particles')


axs[2,1].plot(contrails1.NewIce, label = "Contrails 1")
axs[2,1].plot(contrails2.NewIce, label = "Contrails 2")
axs[2,1].plot(contrails3.NewIce, label = "Contrails 3")
# Set the x axis label of the current axis.
axs[2,1].set_xlabel('Time Steps')
# Set the y axis label of the current axis.
axs[2,1].set_ylabel('Ice')
# Set a title 
axs[2,1].set_title('Total Number of New Ice Particles')

fig.tight_layout()
```
