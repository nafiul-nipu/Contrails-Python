```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
contrails1 = pd.read_csv('Extracted Data/contrails 1.csv')
# contrails1.head()
contrails1.set_index('Timesteps', inplace=True)
contrails1.sort_values(by=["Percentage"], ascending=False)
```
