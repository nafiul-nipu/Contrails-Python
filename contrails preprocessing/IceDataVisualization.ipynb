```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

# for surface plot
from matplotlib import cm
from matplotlib.ticker import LinearLocator
```

```python
contrails1 = pd.read_csv('Extracted Data/contrails 1.csv')
# contrails1.head()
contrails1.set_index('Timesteps', inplace=True)
# contrails1.sort_values(by=["Percentage"], ascending=False)
contrails1
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, figsize=(10,15))
ax1.plot(contrails1.TotalIcePercentage)
ax1.set_title('Data 1')
ax2.plot(contrails2.TotalIcePercentage)
ax2.set_title('Data 2')
ax3.plot(contrails3.TotalIcePercentage)
ax3.set_title('Data 3')

fig.tight_layout()
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

# Make data.
# X = np.arange(-5, 5, 0.25)
X = contrails1.Temp
# Y = np.arange(-5, 5, 0.25)
Y = contrails1.IceVolume
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
# print(Z)

# Z = np.arange(-5, 5, 0.25)

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Customize the z axis.
ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
# A StrMethodFormatter is used automatically
ax.zaxis.set_major_formatter('{x:.02f}')

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
