```python
import pandas as pd
import numpy as np
from os import listdir
import json
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
"column is the attribute name for binning"
column = 'T'
"ranges store all the contrails data range"
ranges = {}
```

```python
"""getting the minimum and maximum range for the binning"""
folders = [f for f in listdir("data/")]
# folders = ['contrails1']
for folder in folders:
    ranges[folder] = list()
    onlyfiles = [f for f in listdir("data/{}".format(folder)) ]
#     onlyfiles = ["0.07.csv", "0.08.csv"]
    path = "data/{}/{}".format(folder, onlyfiles[0])
    timeStepData = pd.read_csv(path)
    iceData = timeStepData[timeStepData.iceLabel == 1]        
    columnData = iceData[column]
    ranges[folder].append(columnData.min())
    ranges[folder].append(columnData.max())
#     print(min_val, max_val)
    for i in range(1, len(onlyfiles)):
        path = "data/{}/{}".format(folder, onlyfiles[i])
        timeStepData = pd.read_csv(path)
        iceData = timeStepData[timeStepData.iceLabel == 1]        
        columnData = iceData[column]
        if columnData.min() < ranges[folder][0] :
            ranges[folder][0] = columnData.min()
        if columnData.max() > ranges[folder][1] : 
            ranges[folder][1] = columnData.max()
        
print(ranges)
```

```python
folders = [f for f in listdir("data/")]
folders = ['contrails1']
fig, axs = plt.subplots(ncols=len(folders))
for folder in folders:
    ranges[folder] = list()
    onlyfiles = [f for f in listdir("data/{}".format(folder)) ]
    for i in range(len(onlyfiles)):
        path = "data/{}/{}".format(folder, onlyfiles[i])
        timeStepData = pd.read_csv(path)
        iceData = timeStepData[timeStepData.iceLabel == 1]        
        columnData = iceData[column]
        sns.kdeplot(columnData)
```
